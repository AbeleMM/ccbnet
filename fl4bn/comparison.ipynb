```python
from typing import cast

from matplotlib.axes import Axes
from pgmpy.utils import get_example_model
from util import BENCHMARK_PIVOT_COL, benchmark_multi


def plot_model_results(model_name: str) -> None:
    model = get_example_model(model_name)
    res = benchmark_multi(
        ref_model=model, nr_clients=4, test_counts=2000, samples_factor=50000, r_seed=42
    )

    for metric in [col for col in res.columns if col != BENCHMARK_PIVOT_COL]:
        pivoted_res = res.pivot(columns=BENCHMARK_PIVOT_COL, values=metric)
        ax = cast(Axes, pivoted_res.plot(kind="bar", title=model_name, ylabel=metric))
```

```python
plot_model_results("asia")
```

```python
plot_model_results("child")
```
